## Exploratory Data Analysis
Checking basic information about the dataset (.head(), .shape, etc.) and the values in our columns. Surface level analysis (identifying correlations)

```python
# importing libraries and functions
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
import sys
sys.path.append('..')

from pandas_profiling import ProfileReport
from scripts import project_functions_v2 as pf 
```

```python
#load dataset
df = pf.load_process_rdata('../../data/raw/SubstanceHarmsData.csv')
```

```python
# correlation matrix (getting a basic understanding of correlations)

correlation = df.corr()

corr_plot = plt.figure()
corr_plot = sns.heatmap(correlation)
corr_plot.set_title("Correlation Matrix")
plt.show()
```

```python
#print(df.head())

fig2 = plt.figure()
fig2 = sns.countplot(data=df,x="Source",hue="Substance")
fig2.legend(labels=["Opioids","Stimulants"])
plt.show()
```

```python
fig3 = plt.figure()
fig3 = sns.displot(df,x="Type of Event",hue="Substance")
fig3.set_xticklabels(rotation=90)
plt.legend(labels=["Opioids","Stimulants"])

plt.show()
```
