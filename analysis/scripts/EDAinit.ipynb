## Exploratory Data Analysis
Checking basic information about the dataset (.head(), .shape, etc.) and the values in our columns. Surface level analysis (identifying correlations)

```python
# importing libraries and functions
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
import sys
sys.path.append('..')

from pandas_profiling import ProfileReport
from scripts import project_functions as pf 
```

```python
#load dataset
df = pf.load_process_data('../../data/raw/SubstanceHarmsData.csv')

#check
df.head(20)
```

```python
# correlation matrix (getting a basic understanding of correlations)
correlation = df.corr()

plt.figure()
corr_plot = sns.heatmap(correlation)
corr_plot.set_title("Correlation Matrix")

# High correlation between Event Category (i.e. type of event) and Substance
plt.show()
```

```python
# bar plot (categorical)

plt.figure()
eda_bar = sns.countplot(y=df["Type of Event"],data=df)
eda_bar.set_title("Categorical Totals")
eda_bar.set_xlabel("Total Occurences")
eda_bar.set_ylabel("Type of Event")
eda_bar.set_xticks([0,100,300,500,1000,1500,2000,3000])
plt.show()
```

```python
plt.figure()
eda_compare = sns.countplot(data=df,y="Type of Event",hue="Substance")
plt.legend(title="Substance", labels=["Opiods","Stimulants"])
plt.show(eda_compare)
```
